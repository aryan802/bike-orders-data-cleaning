### âœ… Step 5: Univariate Analysis
We analyze the distribution of:
- **Numeric Columns** using histograms with KDE
- **Categorical Columns** using count plots


Numeric Columns

```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns

for col in numeric_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

Categorical Columns

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    plt.figure(figsize=(8, 4))

    # If column has more than 15 unique values, show only top 15
    if df[col].nunique() > 15:
        top_values = df[col].value_counts().head(15).index
        sns.countplot(data=df[df[col].isin(top_values)], x=col, order=top_values, palette='Set2')
        plt.title(f'Count of Top 15 {col}')
    else:
        sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette='Set2')
        plt.title(f'Count of {col}')

    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

```

### âœ… Step 6: Bivariate Analysis
Analyze relationships between features:
- Correlation Heatmap for numeric features
- Total Sales by Category
- Price vs Quantity
- Monthly Sales Trend


Correlation Heatmap for numeric features

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

```

Total Sales by Category

```python
df['total_sales'] = df['quantity'] * df['price']
sales_by_category1 = df.groupby('category1')['total_sales'].sum().sort_values(ascending=False)
print("Total Sales by Category1:")
print(sales_by_category1)

# Visualization
plt.figure(figsize=(8, 5))
sales_by_category1.plot(kind='bar', color='steelblue')
plt.title('Total Sales by Category1')
plt.ylabel('Total Sales')
plt.xlabel('Category1')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
sales_by_category2 = df.groupby('category2')['total_sales'].sum().sort_values(ascending=False)
print("\nTotal Sales by Category2:")
print(sales_by_category2)

```

```python
# Visualization
plt.figure(figsize=(10, 5))
sales_by_category2.plot(kind='bar', color='orange')
plt.title('Total Sales by Category2')
plt.ylabel('Total Sales')
plt.xlabel('Category2')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sales_by_combination = df.groupby(['category1', 'category2'])['total_sales'].sum().sort_values(ascending=False)
print("\nTotal Sales by Category1 & Category2:")
print(sales_by_combination)

# Visualization
plt.figure(figsize=(12, 6))
sales_by_combination.unstack().plot(kind='bar', stacked=True, figsize=(12,6))
plt.title('Total Sales by Category1 & Category2')
plt.ylabel('Total Sales')
plt.xlabel('Category1')
plt.xticks(rotation=45)
plt.legend(title='Category2')
plt.tight_layout()
plt.show()
```

### Price vs Quantity
A scatter plot will help visualize if higher-priced bikes tend to sell in smaller quantities.
We will also calculate the correlation between price and quantity.


```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='price', y='quantity', data=df, alpha=0.5)
plt.title('Price vs Quantity')
plt.xlabel('Price')
plt.ylabel('Quantity')
plt.show()

# Correlation
corr = df[['price','quantity']].corr()
print("Correlation between Price & Quantity:\n", corr)
```

### Monthly Sales Trend
We will convert `order.date` to datetime, group by month, and plot total sales per month.
This will show whether sales have seasonal trends or consistent growth.


```python
df['order.date'] = pd.to_datetime(df['order.date'])
sales_by_month = df.groupby(df['order.date'].dt.to_period('M'))['total_sales'].sum()

plt.figure(figsize=(12,6))
sales_by_month.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.ylabel('Total Sales')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

A. Average Price by category1 Ã— category2

 --- Clean & Prep ---

```python
# Convert to datetime first
df = df.copy()  # ensure full copy
df['order.date'] = pd.to_datetime(df['order.date'], errors='coerce')

# Remove 'Unknown'
df = df.loc[df['category1'].str.lower() != 'unknown'].copy()

# Add new columns safely
df = df.assign(
    year=df['order.date'].dt.year,
    month=df['order.date'].dt.to_period('M').astype(str)
)

df.to_csv('/content/drive/My Drive/FDS-Project/data/orders_bikes_cleaned.csv', index=False)
```

# --- A. Average Price by Category1 Ã— Category2 ---

```python
pivot_cat = df.pivot_table(index='category1', columns='category2', values='price', aggfunc='mean')
print("\nAverage Price by Category1 Ã— Category2:\n", pivot_cat)
plt.figure(figsize=(10,6))
sns.heatmap(pivot_cat, annot=True, fmt='.0f', cmap='Blues')
plt.title("Average Price by Category1 Ã— Category2")
plt.tight_layout()
plt.show()
```

# --- B. Average Price by Category1 Ã— Frame ---

```python
pivot_frame = df.pivot_table(index='category1', columns='frame', values='price', aggfunc='mean')
print("\nAverage Price by Category1 Ã— Frame:\n", pivot_frame)

plt.figure(figsize=(8,5))
sns.heatmap(pivot_frame, annot=True, fmt='.0f', cmap='Greens')
plt.title("Average Price by Category1 Ã— Frame")
plt.tight_layout()
plt.show()
```

# --- C. Price Distribution by Category1 ---

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='category1', y='price', data=df)
plt.title("Price Distribution by Category1")
plt.tight_layout()
plt.show()
```

# --- D. Price Distribution by Frame (colored by Category1) ---

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='frame', y='price', hue='category1', data=df)
plt.title("Price by Frame and Category1")
plt.tight_layout()
plt.show()
```

# --- E. Orders Over Time by Category1 ---

```python
monthly_orders = df.groupby(['month','category1']).size().unstack(fill_value=0)
monthly_orders.plot(kind='line', figsize=(12,6), marker='o')
plt.title("Monthly Orders by Category1")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
